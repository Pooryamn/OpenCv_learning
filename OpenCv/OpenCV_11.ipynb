# Apply custom kernels on picture

## kernel 1 :: 3*3 ones with different iterations 
    -> res = cv2.erode(img,kernel,iterations)
## kernel 2 :: opening filter (for deleting noise on an image)
    -> opening = cv2.morphologyEx(noise_image,cv2.MORPH_OPEN,kernel)
    -> this is for pictures with background noise 
## kernel 3 :: closing filter (for deleting noise on an image)
    -> cloing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
    -> this is for pictures with forground noise
## kernel 4 :: gradient filter(edge detection in image)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((300,550))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text="POORYA",org=(20,200),fontFace=font,fontScale=4,color=(255,255,255),thickness = 15)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,20))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
res = cv2.erode(img,kernel,iterations = 1)
```

```python
display_img(res)
```

```python
img = load_img()
```

```python
res = cv2.erode(img,kernel,iterations=3)
```

```python
display_img(res)
```

```python
img = load_img()
```

```python
white_noise = np.random.randint(0,2,size=(300,550))
```

```python
display_img(white_noise)
```

```python
white_noise = white_noise * 255 # 0 , 1 => 0 , 255
```

```python
noise_image = white_noise + img # Write noise on our image
```

```python
display_img(noise_image) # this image has background noise
```

```python
opening = cv2.morphologyEx(noise_image,cv2.MORPH_OPEN,kernel)
```

```python
display_img(opening)
```

```python
img = load_img()
```

```python
black_noise = np.random.randint(low=0,high=2,size=(300,550))
```

```python
black_noise = black_noise * -255
```

```python
display_img(black_noise)
```

```python
black_noise_img = black_noise + img
```

```python
black_noise_img[black_noise == -255] = 0
```

```python
display_img(black_noise_img) # forground noise
```

```python
cloing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
```

```python
display_img(cloing)
```

```python
img = load_img()
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient) # different between closing and opening
```
