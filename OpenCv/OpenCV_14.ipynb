# Histogram equalization

## scale a grayscale image to high contrast image.
### Ex : Before(Darkest = 50,lightest = 200) --> apply Histogram Equalization --> after(darkest = 0,lightest=255)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('horse.jpg') # to work for histogram
horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # to show for me

rainbow_img = cv2.imread('rainbow.jpg')
rainbow = cv2.cvtColor(rainbow_img,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('bricks.jpg')
briks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
img = rainbow_img
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],np.uint8)
```

```python
plt.imshow(mask,cmap='gray')
```

```python
mask[300:400,100:400] = 255
```

```python
plt.imshow(mask,cmap='gray')
```

```python
plt.imshow(rainbow)
```

```python
masked_img = cv2.bitwise_and(img,img,mask=mask)
```

```python
show_mask_img = cv2.bitwise_and(rainbow,rainbow,mask=mask)
```

```python
plt.imshow(show_mask_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow_img],channels=[2],mask=mask,histSize=[256],ranges=[0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow_img],channels=[2],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title('red Hist for masked rainbow')
```

```python
plt.imshow(rainbow)
```

```python
plt.plot(hist_values_red)
plt.title('red Hist for rainbow')
```
