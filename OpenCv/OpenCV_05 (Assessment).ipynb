```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def p_print(img,a,b):
    ax,fig = plt.subplots(figsize=(a,b))
    plt.imshow(img)
```

```python
img = cv2.imread('dog_backpack.jpg')
```

```python
p_print(img,10,10)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
p_print(img,15,15)
```

```python
img2 = cv2.flip(img,0)
```

```python
p_print(img2,10,10)
```

```python
cv2.rectangle(img,pt1=(190,360),pt2=(600,720),color=(255,0,0),thickness=5)
```

```python
p_print(img,10,10)
```

```python
img_tr = img.copy()
```

```python
vertices = np.array([[150,730],[400,350],[650,730]],dtype=np.int32)
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(img_tr,[pts],isClosed=True,color=(0,0,255),thickness=5)
```

```python
p_print(img_tr,10,10)
```

```python
cv2.fillPoly(img_tr,[pts],color=(0,0,255))
```

```python
p_print(img_tr,10,10)
```
