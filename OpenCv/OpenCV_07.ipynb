```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def p_print(img,a,b):
    ax,fig = plt.subplots(figsize=(a,b))
    plt.imshow(img)
```

```python
img1 = cv2.imread('dog_backpack.jpg')
img2 = cv2.imread('watermark_no_copy.png')
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
p_print(img1,10,10)
```

```python
p_print(img2,10,10)
```

```python
img1.shape
```

```python
img2.shape
```

```python
# Resizing ::
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
p_print(img1,10,10)
```

```python
p_print(img2,10,10)
```

```python
blended = cv2.addWeighted(src1=img2,alpha=0.1,src2=img1,beta=1,gamma=0)
```

```python
p_print(blended,10,10)
```

```python
img1 = cv2.imread('dog_backpack.jpg')
img2 = cv2.imread('watermark_no_copy.png')

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2,(600,600))  
```

```python
p_print(img2,10,10)
```

```python
large_img= img1
small_img = img2
```

```python
X_offset = 0
y_offset = 0
```

```python
x_end = X_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,X_offset:x_end] = small_img
```

```python
p_print(large_img,10,10)
```
