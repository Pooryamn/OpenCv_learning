# Histogram:

## Histogram works with BGR order channels so we must pass the orginal opencv image read . not converted to RGB

## to calcute histogram :
    -> cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],range=[0,256]) #  channels : B=0 , G =1 , R=2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('horse.jpg') # to work for histogram
horse = cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) # to show for me

rainbow_img = cv2.imread('rainbow.jpg')
rainbow = cv2.cvtColor(rainbow_img,cv2.COLOR_BGR2RGB)

blue_bricks = cv2.imread('bricks.jpg')
briks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(horse)
```

```python
plt.imshow(rainbow)
```

```python
plt.imshow(briks)
```

```python
hist_vals = cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
hist_vals.shape
```

```python
plt.plot(hist_vals)
```

```python
hist_vals = cv2.calcHist([dark_horse],channels=[0],mask=None,histSize=[256],ranges=[0,256])
```

```python
plt.plot(hist_vals)
```

```python
img = blue_bricks
```

```python
color = ('b','g','r')
```

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])

plt.title('Histograms')
```
