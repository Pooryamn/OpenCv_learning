<h3>
    
           1. imread :: this fuction can have image color theme -> Ex :: cv2.imread('image path',cv2.IMREAD_GRAYSCALE)
    
           2. cvtColor :: Convert source image channel order -> Ex :: cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
           3. resize  -> Ex :: resized_img = cv2.resize(image,(new width pixels , new height pixels)) or resized_img = cv2.resize(image,(0,0),image,widht_ratio,height_ratio) :: ratio = 0.5 = 50% actual size 
    
           4. flip : this fuction is for rotate image -> Ex :: new_img = cv2.flip(image,0) ( 0 : herizontal rotate , 1 : vertical rotate , -1 : herizontal and vertical rotate)
    
           5. imwrite :: this fuction is for save our file -> Ex :: cv2.imwrite('image path',image)
    
           -> if file path is wrong in imread function , we get no errors. instead if we check the image type it is "NoneType" 
    
           -> opencv and matplotlib expect different color channel order (matplotlib expects : RGB but opencv expects : BGR)
    
           ->  for show imag larger in notebook we use fig
    
                    fig = plt.figure(figsize=(10,10))
                    ax = add_subplot(111) -> 1 * 1 * 1 image 
                    ax.imshow(image)
        
</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(img)
```

```python
# Matplotlib --> RGB
# OpenCV --> BGR
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(img)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
img_gray = cv2.cvtColor(img_gray,cv2.COLOR_BGR2RGB)
```

```python
img_gray.shape
```

```python
img_gray.min()
```

```python
img_gray.max()
```

```python
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(img_gray)
```

```python
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(img,cmap='gray')
```

```python
plt.imshow(img)
```

```python
resize_img = cv2.resize(img,(400,1000))
```

```python
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(resize_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_imf = cv2.flip(img,-1)#miror # -1 0 1
fix, ax = plt.subplots(figsize = (10,15))
plt.imshow(new_imf)
```

```python
cv2.imwrite('test.jpg',new_imf)
```

```python
out_img = cv2.cvtColor(new_imf,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(out_img)
```

```python
cv2.imwrite('test2.jpg',out_img)
```
